Making Changes in the data

```python
pokemon_data['Total'] = pokemon_data['HP'] + pokemon_data['Attack'] + pokemon_data['Defense'] + pokemon_data['Sp. Atk'] + pokemon_data['Sp. Def'] + pokemon_data['Speed']
pokemon_data.head()
```

```python
# Dropping or Deleting a Column from database
pokemon_data = pokemon_data.drop(columns=['Total'])
pokemon_data
```

```python
pokemon_data['Total'] = pokemon_data.iloc[:, 4:10].sum(axis=1)

pokemon_data.head()
```

```python
cols = list(pokemon_data.columns.values)

# We are changing the position to the column "Total"
pokemon_data = pokemon_data[cols[0:4] + [cols[-1]] + cols[4:12]]
pokemon_data
```

```python
# Saving or Creating a new csv file
# You can create .xlsx , .txt , .csv
pokemon_data.to_csv("modified.txt" , index=False)  

pokemon_data.to_excel('modified.xlsx' , index=False)

pokemon_data.to_csv('modified.csv',index=False)
```

```python
# Filtering Data
# Showing only rows which has grass.

new_df = pokemon_data.loc[(pokemon_data['Type 1'] == 'Grass') & (pokemon_data['Type 2'] == "Poison") & (pokemon_data['HP'] > 70)]

# Resetting the indices of the rows after selecting data from the main csv file
new_df = new_df.reset_index()
# OR
new_df.reset_index(drop=True , inplace =True)
```

```python
# This willhelps us to get names containing "Mega"
pokemon_data.loc[pokemon_data['Name'].str.contains('Mega')]

# This ~ will removes all the names containing 'Mega'
pokemon_data.loc[~pokemon_data['Name'].str.contains('Mega')]
```

```python
# Regex Filtering
import re

# pokemon_data.loc[pokemon_data['Type 1'].str.contains('Fire|Grass', flags=re.I , regex=True)]

pokemon_data.loc[pokemon_data['Name'].str.contains('^pi[a-z]*', flags=re.I , regex=True)]
```

```python
# Conditional Changes
# Converting Fire into Flamer
pokemon_data.loc[pokemon_data['Type 1'] == 'Fire' , 'Type 1'] = 'Flamer'
pokemon_data
```

```python
# Aggegarte Statistics (GroupBy)
# sum , mean , count
df = pd.read_csv('pokemon_data.csv')

df.groupby(['Type 1']).mean().sort_values('Attack' , ascending=False)
```

```python
df['count'] = 1

df.groupby(['Type 1' , 'Type 2']).count()['count']
```
